```python
from scipy.spatial import distance
gen, org = pickle.load(open('middle.pkl', 'rb'))

to_json = []
for i, (g, o) in enumerate(zip(gen, org)):
    if g:
        sim = distance.cosine(g['embedding']['vector'], o['embedding']['vector'])
        to_json.append({
            "sim": 1-sim,
            "gen":{
                "title": g["title"],
                "url": g["url"],
                'abstract': g['abstract'],
                'fieldsOfStudy': g['fieldsOfStudy'],
            },
            "org":{
                "title": o["title"],
                "url": o["url"],
                'abstract': o['abstract'],
                'fieldsOfStudy': o['fieldsOfStudy'],
            }
        })
    else:
        to_json.append(None)

json.dump(to_json, open("result/middle.json", 'w'), indent=2)
    
```
